```python
import pandas as pd
import numpy as np

# Укажите правильный путь к файлу
file_path = "data/marketing_campaign.xlsx"

# Загрузите файл
df = pd.read_excel(file_path)
df.head(10)
```

```python
df['Income'] = df['Income'].fillna(df['Income'].median())  # Заменить на медиану
```

```python
# Check for missing data
print("Missing data in the dataframe:")
print(df.isnull().sum())
```

```python
# Feature Engineer a breakpoint
print("Unique values in Education column:", df["Education"].unique())
print("Unique values in Martial Status column:", df["Marital_Status"].unique())
```

```python
def education_level(education):
    if education in ["Graduation","phD","Master"]:
        return "high"
    elif education in ["Basic"]:
        return "Middle"
    else:
        return "Low"

df["Education_level"] = df["Education"].apply(education_level)
```

```python
def living_status(marital_status):
    if marital_status in ['Alone', 'Absurd', 'YOLO']:
        return 'Living Alone'
    else:
        return 'Living with Others'

df['Living_Status'] = df['Marital_Status'].apply(living_status)
```

```python
df["Age"] = 2024 - df["Year_Birth"]
```

```python
df['Total_Campaigns_Accepted'] = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].sum(axis=1)
```

```python
df['Average_Spend'] = (df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)) / df['NumDealsPurchases']
```

```python
df['Spent'] = df['MntWines']+df["MntWines"] +df['MntFruits']+ df['MntMeatProducts'] +df['MntFishProducts']+df['MntSweetProducts']+ df['MntGoldProds']
```

```python
df['Is_Parent'] = (df['Kidhome'] + df['Teenhome'] > 0).astype(int)
```

```python
# create new feature for total spending
df['total_spending'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']

# create new feature for average monthly visits to the company's website
df['avg_web_visits'] = df['NumWebVisitsMonth'] / 12

# create new feature for the ratio of online purchases to total purchases
df['online_purchase_ratio'] = df['NumWebPurchases'] / (df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases'])
```

```python
to_drop = ['Dt_Customer', 'Z_CostContact', 'Z_Revenue', 'Year_Birth', 'ID']
df = df.drop(to_drop, axis=1)
```

```python
import plotly.express as px
import pandas as pd

# Основная функция для анализа трат по категориям продуктов по уровням образования
def analyze_spending_by_education_and_product(df):
    # Заменим инфини и NaN значения на 0 для столбца 'Average_Spend'
    df['Average_Spend'].replace([np.inf, -np.inf], np.nan, inplace=True)
    df['Average_Spend'].fillna(0, inplace=True)

    # Убедитесь, что столбец 'Education_level' существует
    if 'Education_level' not in df.columns:
        raise ValueError("The dataframe must contain 'Education_level' column")

    # Группировка данных по уровню образования и вычисление среднего значения трат для каждого продукта
    product_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

    df_grouped = df.groupby('Education_level')[product_columns].mean().reset_index()

    # Создание графика
    fig = px.bar(df_grouped, x='Education_level', y=product_columns,
                 title="Average Spending by Education Level on Different Products",
                 labels={'Education_level': 'Education Level',
                         'value': 'Average Spend',
                         'variable': 'Product'},
                 barmode='group')

    # Подписи для оси X и Y
    fig.update_layout(
        xaxis_title="Education Level",
        yaxis_title="Average Spend",
        xaxis_tickmode="array",
        xaxis_tickvals=[0, 1, 2],  # Подгоняем метки для категорий (Low, Middle, High)
        xaxis_ticktext=['Low', 'Middle', 'High']
    )

    # Сохраняем график в HTML файл
    file_path = '/content/spending_by_education_and_product.html'
    fig.write_html(file_path)

    # Скачиваем файл
    from google.colab import files
    files.download(file_path)

    return file_path

# Пример использования функции
file_path = analyze_spending_by_education_and_product(df)
```

```python
import plotly.express as px
import pandas as pd

# Основная функция для анализа уровня доходов по категориям образования
def analyze_income_by_education(df):
    # Убедитесь, что столбец 'Income' существует
    if 'Income' not in df.columns:
        raise ValueError("The dataframe must contain 'Income' column")

    # Группируем данные по уровню образования и вычисляем средний доход для каждой категории
    df_grouped_income = df.groupby('Education_level')['Income'].mean().reset_index()

    # Создание графика
    fig = px.bar(df_grouped_income, x='Education_level', y='Income',
                 title="Average Income by Education Level",
                 labels={'Education_level': 'Education Level', 'Income': 'Average Income'},
                 color='Education_level')

    # Подписи для оси X и Y
    fig.update_layout(
        xaxis_title="Education Level",
        yaxis_title="Average Income",
        xaxis_tickmode="array",
        xaxis_tickvals=[0, 1, 2],  # Подгоняем метки для категорий (Low, Middle, High)
        xaxis_ticktext=['Low', 'Middle', 'High']
    )

    # Сохраняем график в HTML файл
    file_path = '/content/income_by_education_level.html'
    fig.write_html(file_path)

    # Скачиваем файл
    from google.colab import files
    files.download(file_path)

    return file_path

# Пример использования функции
file_path_income = analyze_income_by_education(df)
```

```python
import plotly.express as px
import pandas as pd

# Функция для построения и скачивания графиков зависимости количества кампаний от дохода
def plot_and_download_campaigns_by_income(df):
    # Создаем новый столбец с диапазонами доходов для группировки
    df['Income_Range'] = pd.cut(df['Income'], bins=6)

    # Сортируем интервалы по возрастанию
    df['Income_Range'] = df['Income_Range'].astype(str)

    # Группируем по диапазону дохода и находим среднее количество принятых кампаний
    df_grouped = df.groupby('Income_Range')['Total_Campaigns_Accepted'].mean().reset_index()

    # Преобразуем интервалы в категорию и упорядочиваем их
    df_grouped['Income_Range'] = pd.Categorical(df_grouped['Income_Range'],
                                                categories=sorted(df['Income_Range'].unique(), key=lambda x: float(x.split(',')[0][1:])) ,
                                                ordered=True)

    # Сортируем DataFrame
    df_grouped = df_grouped.sort_values('Income_Range')

    # Строим столбчатую диаграмму
    fig = px.bar(
        df_grouped,
        x='Income_Range',
        y='Total_Campaigns_Accepted',
        title='Total Campaigns Accepted vs Income (Bar Chart)',
        labels={'Income_Range': 'Income Range', 'Total_Campaigns_Accepted': 'Average Total Campaigns Accepted'},
        text='Total_Campaigns_Accepted',  # Добавляем подписи
        color='Total_Campaigns_Accepted'  # Цветовая индикация по количеству принятых кампаний
    )

    # Настроим формат оси X для показа интервалов доходов
    fig.update_layout(
        xaxis_title='Income Range',
        yaxis_title='Average Total Campaigns Accepted',
    )

    # Отображаем график
    fig.show()

# Пример использования функции
plot_and_download_campaigns_by_income(df)
```

```python
import plotly.express as px
import pandas as pd

# Функция для построения и скачивания графиков зависимости количества кампаний от возраста, разделенного на 3 группы
def plot_and_download_campaigns_by_age_grouped(df):
    # Разбиваем возраст на 3 группы
    age_bins = [0, 30, 50, 100]  # Примерные возрастные группы: 0-30, 31-50, 51-100
    age_labels = ['18-30', '31-50', '51+']  # Метки для групп

    # Создаем новый столбец с категориями возраста
    df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

    # Группируем данные по возрастным группам и находим среднее количество принятых кампаний
    df_grouped = df.groupby('Age_Group')['Total_Campaigns_Accepted'].mean().reset_index()

    # Создаем столбчатую диаграмму
    fig = px.bar(
        df_grouped,
        x='Age_Group',
        y='Total_Campaigns_Accepted',
        title='Total Campaigns Accepted by Age Group (Bar Chart)',
        labels={'Age_Group': 'Age Group', 'Total_Campaigns_Accepted': 'Average Total Campaigns Accepted'},
        text='Total_Campaigns_Accepted',  # Добавляем подписи
        color='Age_Group'  # Цветовая индикация по возрастной группе
    )

    # Настроим формат оси X для показа возрастных групп
    fig.update_layout(
        xaxis_title='Age Group',
        yaxis_title='Average Total Campaigns Accepted',
    )

    # Отображаем график
    fig.show()

# Пример использования функции
plot_and_download_campaigns_by_age_grouped(df)
```

```python
import plotly.express as px
import pandas as pd
from google.colab import files

# Функция для построения и скачивания графиков распределения трат по доходу
def plot_and_download_spending_by_income(df):
    # Создаем новый столбец с диапазонами доходов для группировки
    df['Income_Range'] = pd.cut(df['Income'], bins=6)

    # Сортируем интервалы по возрастанию
    df['Income_Range'] = df['Income_Range'].astype(str)
    df_grouped = df.groupby('Income_Range')['Total_Spend'].mean().reset_index()

    # Преобразуем интервалы в категорию и упорядочиваем их
    df_grouped['Income_Range'] = pd.Categorical(df_grouped['Income_Range'],
                                                categories=sorted(df['Income_Range'].unique(), key=lambda x: float(x.split(',')[0][1:])),
                                                ordered=True)

    # Сортируем DataFrame
    df_grouped = df_grouped.sort_values('Income_Range')

    # Строим столбчатую диаграмму
    fig = px.bar(
        df_grouped,
        x='Income_Range',
        y='Total_Spend',
        title='Total Spend vs Income (Bar Chart)',
        labels={'Income_Range': 'Income Range', 'Total_Spend': 'Average Total Spend'},
        text='Total_Spend',  # Добавляем подписи
        color='Total_Spend'  # Цветовая индикация по сумме трат
    )

    # Настроим формат оси X для показа интервалов доходов
    fig.update_layout(
        xaxis_title='Income Range',
        yaxis_title='Average Total Spend',
    )


    # Отображаем график
    fig.show()

# Пример использования функции
plot_and_download_spending_by_income(df)
```

```python
import plotly.express as px

# Функция для построения и скачивания графиков распределения трат
def plot_and_download_spending_distributions(df):
    # График: Распределение трат по семейному статусу
    fig11 = px.bar(
        df.groupby('Marital_Status')['Total_Spend'].sum().reset_index(),  # Группируем по семейному статусу
        x='Marital_Status',  # Семейный статус на оси X
        y='Total_Spend',  # Сумма трат на оси Y
        color='Marital_Status',  # Цвет для разных семейных статусов
        title='Spending Distribution by Marital Status',
        labels={'Marital_Status': 'Marital Status', 'Total_Spend': 'Total Spending'}
    )

    # Отображаем график
    fig11.show()

# Пример использования функции
plot_and_download_spending_distributions(df)
```
